```python
# import libraries
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
from selenium import webdriver
```

```python
def remove_tags(s):
    s = s.replace('<title>', '').replace('</title>', '').replace('<link>', '').replace('</link>', '')
    return s
```

```python
class itemClass:
    def __init__(self, title, link, content):
        self.title = title
        self.link = link
        self.content = content

    def __json__(self):
        # Define a method to convert your object to a JSON-serializable format
        return {
            'title': self.title,
            'link': self.link,
            'content': self.content,
        }
```

```python
#Loop to go over all pages
pages = np.arange(1, 3, 1)
data=[]
content_data = ""

for page in pages:
    
    page="https://rss.app/feeds/XenZCYoo7eqBLeHy.xml" + str(page) 
    driver = webdriver.Chrome()
    driver.get(page)  
    sleep(randint(2,10))
    soup = BeautifulSoup(driver.page_source, 'xml')
    my_table = soup.find_all(['item'])

    for item in my_table:
        title = item('title')
        link = item('link')

        obj = itemClass(remove_tags(str(title)), remove_tags(str(link)), content_data)
        data.append(obj)
```

```python
import json

json_array = json.dumps([{'title': item.title, 'link': item.link, 'content': item.content} for item in data], indent=2)

# Print the JSON array
print(json_array)
```

```python
title = ""
link = ""
final_result = []
json_array_str = json.loads(json_array)
for item in json_array_str:
    print(item['link'])
    title = item['title']
    link = str(item['link'].replace("[", "").replace("]", ""))
    #Loop to go over all pages
    pages = np.arange(1)
    content_data=[]
    
    for page in pages:
        
        page= link + str(page) 
        driver = webdriver.Chrome()
        driver.set_page_load_timeout(300)
        driver.get(page)  
        sleep(randint(2,10))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        my_table = soup.find_all(['article'])
    
        for item in my_table:
            content = item(lambda tag: tag.name == 'p' and not tag.attrs)
            content_data.append(content)

        flat_list = [str(item) for sublist in content_data for item in sublist]
    
        # Create a single string using join
        result_string = ' '.join(flat_list)
        
        # Parse the HTML string
        soup = BeautifulSoup(result_string, 'html.parser')
        
        # Get the text content without HTML tags
        text_without_tags = soup.get_text()

    obj = itemClass(title, link, text_without_tags)
    final_result.append(obj)
```

```python
import json

json_array_1 = json.dumps([{'title': item.title, 'link': item.link, 'content': item.content} for item in final_result], indent=2)

# Print the JSON array
print(json_array_1)
```

```python
import json

json_array = json.dumps([{'title': item.title, 'link': item.link, 'content': item.content} for item in final_result], indent=2)

# Print the JSON array
print(json_array)
```

```python
import csv

# Specify the CSV file name
csv_file = 'output.csv'

# Writing to CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    # Get the keys dynamically from the first item in the array
    fieldnames = json.loads(json_array)[0].keys()
    
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write the header
    writer.writeheader()
    
    # Write the data
    writer.writerows(json.loads(json_array))

print(f'CSV file "{csv_file}" has been created successfully.')
```
